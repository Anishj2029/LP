```python
import tkinter as tk

# =========================
# KNOWLEDGE BASE
# =========================
class KnowledgeBase:

    def __init__(self):

        self.rules = [

            {
                "keywords": ["database", "data missing"],
                "solution": "Check database connection and restore backup if needed."
            },

            {
                "keywords": ["network", "slow internet"],
                "solution": "Restart router and check network configuration."
            },

            {
                "keywords": ["server", "crash"],
                "solution": "Restart the server and monitor CPU/RAM usage."
            },

            {
                "keywords": ["security", "unauthorized access"],
                "solution": "Change passwords and enable firewall protection."
            },

            {
                "keywords": ["backup", "failure"],
                "solution": "Verify backup logs and configure automatic backups."
            },

            {
                "keywords": ["software", "installation"],
                "solution": "Reinstall or update the information management software."
            },

            {
                "keywords": ["performance", "slow"],
                "solution": "Close unnecessary applications and optimize database queries."
            }
        ]


# =========================
# KNOWLEDGE ACQUISITION
# =========================
class KnowledgeAcquisition:

    def add_rule(self, kb, keywords, solution):

        kb.rules.append({

            "keywords": keywords,
            "solution": solution
        })


# =========================
# INFERENCE ENGINE
# =========================
class InferenceEngine:

    def infer(self, kb, user_input):

        best_match = None
        max_score = 0

        for rule in kb.rules:

            score = sum(

                1 for word in rule["keywords"]

                if word in user_input
            )

            if score > max_score:

                max_score = score
                best_match = rule

        if best_match:

            return best_match["solution"], max_score

        return "Issue unclear. Please contact IT administrator.", 0


# =========================
# EXPLANATION MODULE
# =========================
class ExplanationModule:

    def explain(self, score):

        if score > 0:

            return f"Matched {score} keyword(s) from knowledge base."

        return "No relevant rule matched."


# =========================
# EXPERT SYSTEM
# =========================
class ExpertSystem:

    def __init__(self):

        self.kb = KnowledgeBase()

        self.ie = InferenceEngine()

        self.explainer = ExplanationModule()

        self.ka = KnowledgeAcquisition()

    def process(self, user_input):

        solution, score = self.ie.infer(

            self.kb,
            user_input
        )

        explanation = self.explainer.explain(score)

        return solution, explanation


# Create system object
system = ExpertSystem()


# =========================
# SEND MESSAGE FUNCTION
# =========================
def send_message(event=None):

    user_input = entry.get().lower().strip()

    if not user_input:
        return

    chat_box.config(state=tk.NORMAL)

    chat_box.insert(
        tk.END,
        "👤 You: " + user_input + "\n"
    )

    entry.delete(0, tk.END)

    # Exit option
    if user_input == "exit":

        chat_box.insert(
            tk.END,
            "🤖 System: Session ended. Thank you!\n"
        )

        chat_box.config(state=tk.DISABLED)

        return

    # Help option
    if "help" in user_input:

        chat_box.insert(
            tk.END,

            "🤖 You can ask about:\n"
            "- database issues\n"
            "- network problems\n"
            "- server crash\n"
            "- backup failure\n"
            "- software installation\n"
            "- security issues\n\n"
        )

        chat_box.config(state=tk.DISABLED)

        return

    # Process input
    solution, explanation = system.process(user_input)

    chat_box.insert(tk.END, "🤖 System:\n")

    chat_box.insert(
        tk.END,
        "✔️ Solution: " + solution + "\n"
    )

    chat_box.insert(
        tk.END,
        "🧠 Reason: " + explanation + "\n\n"
    )

    chat_box.config(state=tk.DISABLED)

    chat_box.yview(tk.END)


# =========================
# GUI
# =========================
root = tk.Tk()

root.title("Information Management Expert System")

root.geometry("500x550")


# Chat box
chat_box = tk.Text(

    root,
    font=("Arial", 11),
    wrap=tk.WORD
)

chat_box.pack(

    padx=10,
    pady=10,
    fill=tk.BOTH,
    expand=True
)

chat_box.config(state=tk.DISABLED)


# Welcome message
chat_box.config(state=tk.NORMAL)

chat_box.insert(

    tk.END,

    "🤖 Welcome to the Information Management Expert System\n"
    "--------------------------------------------------------------\n"
    "Type your issue below 👇 or type 'help' for options.\n"
    "Type 'exit' to quit.\n\n"
)

chat_box.config(state=tk.DISABLED)


# Input frame
frame = tk.Frame(root)

frame.pack(

    padx=10,
    pady=5,
    fill=tk.X
)


# Entry box
entry = tk.Entry(

    frame,
    font=("Arial", 12)
)

entry.pack(

    side=tk.LEFT,
    fill=tk.X,
    expand=True,
    padx=(0, 5)
)

entry.bind("<Return>", send_message)


# Send button
send_btn = tk.Button(

    frame,
    text="Send",
    command=send_message
)

send_btn.pack(side=tk.RIGHT)


# Run GUI
root.mainloop()
```
