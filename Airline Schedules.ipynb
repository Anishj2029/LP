```python
import tkinter as tk

# =========================
# KNOWLEDGE BASE
# =========================
class KnowledgeBase:

    def __init__(self):

        self.rules = [

            {
                "keywords": ["flight delay", "weather"],
                "solution": "Flight delayed due to bad weather conditions. Reschedule passengers."
            },

            {
                "keywords": ["cargo", "late shipment"],
                "solution": "Cargo shipment delayed. Arrange alternate transportation."
            },

            {
                "keywords": ["technical issue", "aircraft"],
                "solution": "Aircraft requires maintenance inspection before departure."
            },

            {
                "keywords": ["overbooking", "seat"],
                "solution": "Provide passengers with alternate flights or compensation."
            },

            {
                "keywords": ["fuel", "shortage"],
                "solution": "Refueling required before flight scheduling."
            },

            {
                "keywords": ["crew", "unavailable"],
                "solution": "Assign backup crew members for the flight."
            },

            {
                "keywords": ["customs", "documents"],
                "solution": "Verify cargo customs documents before shipment."
            },

            {
                "keywords": ["airport congestion", "delay"],
                "solution": "Adjust flight timings due to airport traffic congestion."
            }
        ]


# =========================
# KNOWLEDGE ACQUISITION
# =========================
class KnowledgeAcquisition:

    def add_rule(self, kb, keywords, solution):

        kb.rules.append({

            "keywords": keywords,
            "solution": solution
        })


# =========================
# INFERENCE ENGINE
# =========================
class InferenceEngine:

    def infer(self, kb, user_input):

        matched_solutions = []
        total_score = 0

        for rule in kb.rules:

            score = sum(
                1 for word in rule["keywords"]
                if word in user_input
            )

            if score > 0:

                matched_solutions.append(
                    rule["solution"]
                )

                total_score += score

        if matched_solutions:

            final_solution = "\n".join(

                f"- {solution}"

                for solution in matched_solutions
            )

            return final_solution, total_score

        return "Issue unclear. Please contact human support.", 0

# =========================
# EXPLANATION MODULE
# =========================
class ExplanationModule:

    def explain(self, score):

        if score > 0:

            return f"Matched {score} keyword(s) from airline scheduling knowledge base."

        return "No relevant airline scheduling rule matched."


# =========================
# EXPERT SYSTEM
# =========================
class ExpertSystem:

    def __init__(self):

        self.kb = KnowledgeBase()

        self.ie = InferenceEngine()

        self.explainer = ExplanationModule()

        self.ka = KnowledgeAcquisition()

    def process(self, user_input):

        solution, score = self.ie.infer(

            self.kb,
            user_input
        )

        explanation = self.explainer.explain(score)

        return solution, explanation


# Create system object
system = ExpertSystem()
system.ka.add_rule(

    system.kb,

    ["wifi", "disconnect"],

    "Restart WiFi adapter and check signal strength."
)



# =========================
# SEND MESSAGE FUNCTION
# =========================
def send_message(event=None):

    user_input = entry.get().lower().strip()

    if not user_input:
        return

    chat_box.config(state=tk.NORMAL)

    chat_box.insert(
        tk.END,
        "👤 Operator: " + user_input + "\n"
    )

    entry.delete(0, tk.END)

    # Exit option
    if user_input == "exit":

        chat_box.insert(
            tk.END,
            "✈️ Airline System: Session ended successfully.\n"
        )

        chat_box.config(state=tk.DISABLED)

        return

    # Help option
    if "help" in user_input:

        chat_box.insert(
            tk.END,

            "✈️ You can ask about:\n"
            "- flight delays\n"
            "- cargo shipment issues\n"
            "- aircraft technical problems\n"
            "- overbooking issues\n"
            "- fuel shortage\n"
            "- customs verification\n"
            "- crew availability\n"
            "- airport congestion\n\n"
        )

        chat_box.config(state=tk.DISABLED)

        return

    # Process input
    solution, explanation = system.process(user_input)

    chat_box.insert(tk.END, "✈️ Airline Scheduling Expert System:\n")

    chat_box.insert(
        tk.END,
        "✔️ Solution: " + solution + "\n"
    )

    chat_box.insert(
        tk.END,
        "🧠 Reason: " + explanation + "\n\n"
    )

    chat_box.config(state=tk.DISABLED)

    chat_box.yview(tk.END)


# =========================
# GUI
# =========================
root = tk.Tk()

root.title("Airline Scheduling and Cargo Expert System")

root.geometry("600x550")


# Chat box
chat_box = tk.Text(

    root,
    font=("Arial", 11),
    wrap=tk.WORD
)

chat_box.pack(

    padx=10,
    pady=10,
    fill=tk.BOTH,
    expand=True
)

chat_box.config(state=tk.DISABLED)


# Welcome message
chat_box.config(state=tk.NORMAL)

chat_box.insert(

    tk.END,

    "✈️ Welcome to Airline Scheduling and Cargo Expert System\n"
    "-----------------------------------------------------------------\n"
    "Enter airline or cargo scheduling issue below 👇\n"
    "Type 'help' for options.\n"
    "Type 'exit' to quit.\n\n"
)

chat_box.config(state=tk.DISABLED)


# Input frame
frame = tk.Frame(root)

frame.pack(

    padx=10,
    pady=5,
    fill=tk.X
)


# Entry box
entry = tk.Entry(

    frame,
    font=("Arial", 12)
)

entry.pack(

    side=tk.LEFT,
    fill=tk.X,
    expand=True,
    padx=(0, 5)
)

entry.bind("<Return>", send_message)


# Send button
send_btn = tk.Button(

    frame,
    text="Send",
    command=send_message
)

send_btn.pack(side=tk.RIGHT)


# Run GUI
root.mainloop()
```
